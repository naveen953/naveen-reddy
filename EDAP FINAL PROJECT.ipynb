### Univariate Analysis for Numerical data

```python
# Histogram for numerical data
numerical_cols = ['Row ID', 'Order ID', 'Order Date', 'Ship Date', 'Customer ID','Postal Code','Sales','Quantity','Discount','Profit']
df[numerical_cols].hist(bins=15, figsize=(25, 10), layout=(3, 3));
```

```python
# Boxplot for numerical data
df[numerical_cols].boxplot()
plt.xticks(rotation=45)
```

```python
#Graphical representation for sales
plt.figure(figsize=(15, 5))
plt.title('Box Plot', fontsize=18)
plt.xlabel('xlabel',fontsize=18)
sns.boxplot(df['Sales'])
```

#### The above representation is the boxplot for the sales of the supermarket. The boxplot is used for the clear interpretation of the outlier and in the picture we ca easily visualize that, most of the sales lying between the interval of 0 to 200.

## Bivariate Analysis

```python
df.plot(x="Region", y="Sales", kind="scatter", figsize=(10, 6), title="Regional Sales", color="salmon");
```

```python
df.corr()
```

```python
corr = df.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr, cmap='mako_r',annot=True)
plt.show()
```

```python
#Graphical representation for sales and sub-category
plt.figure(figsize = (12,4))
sns.barplot(x="Sub-Category",y="Sales", data=df, palette = "Spectral");
plt.xticks(rotation = 90)
plt.show()
```

```python
#Graphical Representation for profit and sales in sub-category
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Sub-Category')
# plt.legend('Profit')
# plt.legend('Sales')
plt.show()
```

```python
#Graphical Representation for discounts in different categories
df.groupby('Sub-Category').mean()['Discount'].plot(kind='line',figsize=(20,10))
plt.ylabel("Discount")
plt.title("Discounts of different categorys")
plt.show()
```

## MULTIVARIATE ANALYSIS

### PAIRPLOT

```python
sns.pairplot(df)
plt.show()
```

```python
sns.pairplot(
data=df,
aspect=.85,
hue="Ship Mode")
```

```python
df.corr()
```

```python
sns.set(font_scale=1.15)
plt.figure(figsize=(8,4))
sns.heatmap(
    df.corr(),        
    cmap='RdBu_r', 
    annot=True, 
    vmin=-1, vmax=1);
```

```python
#Top 20 Customers who benefitted the store
sortedTop20 = df.sort_values(['Profit'], ascending=False).head(20)
fig = plt.figure(figsize=(25,15))
ax = fig.add_subplot(111)
p = sns.barplot(x='Customer Name', y='Profit',hue='State',palette='Set1', data=sortedTop20, ax=ax)
ax.set_title("Top 20 profitable Customers")
ax.set_xticklabels(p.get_xticklabels(), rotation=75)
plt.tight_layout()
plt.show()
```

```python
#Sales per year
df.groupby('order year')['Sales','Profit'].agg(['sum']).plot.bar()
plt.title('Year wise Total Sales & profit gained')
```
